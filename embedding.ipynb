##BERT for F100 !!!

```python
import pandas as pd
from transformers import BertTokenizer, BertModel
import torch
import numpy as np

# 注转 
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# 注转 转
df = pd.read_pickle("/content/drive/My Drive/Colab_Notebooks/data/full_embedding_15K.pkl")

# 驻 转 注转 F100 - 驻转  注专 专转 驻转 NaN
df['F100'] = df['F100'].astype(str).fillna("")

def get_bert_embedding(text, index, total):
    """ 驻拽爪 砖爪专转 Embeddings 注 驻 砖转 拽住 """
    if pd.isna(text) or not isinstance(text, str) or text.strip() == "":
        return np.zeros(768)  # 专 拽专 驻住   拽住 转拽

    if index % 500 == 0:
        print(f"Processing title {index}/{total}...")

    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=100)
    with torch.no_grad():
        output = model(**inputs)
    return output.last_hidden_state[:, 0, :].squeeze().numpy()

#  爪'拽
chunk_size = 500

# 注 转 爪'拽
for i in range(0, len(df), chunk_size):
    chunk = df.iloc[i:i + chunk_size].copy()  # 砖砖 -copy() 注转 SettingWithCopyWarning

    # 爪专转 Embeddings
    chunk['bert_F100_embedding'] = [get_bert_embedding(text, j, len(chunk)) for j, text in enumerate(chunk['F100'])]

    # 注 转 专 -DataFrame 专砖
    df.loc[chunk.index, 'bert_F100_embedding'] = chunk['bert_F100_embedding']

    print(f"Processed chunk {i // chunk_size + 1} of {len(df) // chunk_size + 1}")

# 砖专转 转
df.to_pickle("/content/drive/My Drive/Colab_Notebooks/data/full_embedding_15K.pkl")
print(" Embeddings 砖专 爪!")
```

```python
import pandas as pd
from transformers import BertTokenizer, BertModel
import torch
import numpy as np

# 注转 
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# 驻 转 注转 F100 - 驻转  注专 专转 驻转 NaN
df['F100'] = df['F100'].astype(str).fillna("")

def get_bert_embedding(text, index, total):
    """ 驻拽爪 砖爪专转 Embeddings 注 驻 砖转 拽住 """
    if pd.isna(text) or not isinstance(text, str) or text.strip() == "":
        return np.zeros(768)  # 专 拽专 驻住   拽住 转拽

    if index % 500 == 0:
        print(f"Processing title {index}/{total}...")

    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=100)
    with torch.no_grad():
        output = model(**inputs)
    return output.last_hidden_state[:, 0, :].squeeze().numpy()

# 专转 住驻专 砖专转 注
num_rows_to_process = 5000

# 注 转 注专 5000 砖专转 专砖转 
chunk = df.iloc[:num_rows_to_process].copy()  # 砖砖 -copy() 注转 SettingWithCopyWarning

# 爪专转 Embeddings
chunk['bert_F100_embedding'] = [get_bert_embedding(text, j, len(chunk)) for j, text in enumerate(chunk['F100'])]

# 注 转 专 -DataFrame 专砖
df.loc[chunk.index, 'bert_F100_embedding'] = chunk['bert_F100_embedding']

print(f"Processed {num_rows_to_process} rows.")

# 砖专转 转
df.to_pickle("/content/drive/My Drive/Colab_Notebooks/data/full_embedding_15K.pkl")
print(" Embeddings 砖专 爪!")
```
