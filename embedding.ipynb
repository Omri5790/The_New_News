## 15K embeddings


```python
import torch
from transformers import BertTokenizer, BertModel
from sentence_transformers import SentenceTransformer
import pandas as pd
#נטען דאטה

file_path = "/content/drive/My Drive/Colab_Notebooks/data/cleaned_15k.csv"
df = pd.read_csv(file_path)


```

##Agnostic embedding titles

```python
from sentence_transformers import SentenceTransformer

agnostic_model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

# פונקציה להפקת Embeddings לכותרות עם הדפסה למעקב התקדמות
def get_agnostic_embedding(text, index, total):
    if index % 500 == 0:  # הדפסה כל 500 שורות
        print(f"📌 Processing title {index}/{total}...")

    return agnostic_model.encode(text)

# הפעלת הפונקציה על כל הכותרות עם מעקב
total_titles = len(df)
df['agnostic_title_embedding'] = [get_agnostic_embedding(text, i, total_titles) for i, text in enumerate(df['title'])]

# שמירת ה-embeddings לשימוש עתידי
df.to_pickle("/content/drive/My Drive/Colab_Notebooks/data/full_embedding_15K.pkl")

print(" Agnostic BERT  title embeddings create")
```

#BERT for titles


```python

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

#embeddinf func
def get_bert_embedding(text, index, total):
    if index % 500 == 0:  # הדפסה כל 500 שורות
        print(f"Processing title {index}/{total}...")

    input = tokenizer(text, return_tensors='pt', truncation=True, padding=True)
    with torch.no_grad():
      output = model(**input)
    return output.last_hidden_state[:, 0,:].squeeze().numpy()#ייצוג של כל המשפט , ולא רק של מילה


# שמירה בקובץ פיקל
total_titles = len(df)
df['bert_embedding'] = [get_bert_embedding(text, i, total_titles) for i, text in enumerate(df['title'])]

df.to_pickle("/content/drive/My Drive/Colab_Notebooks/data/full_embedding_15K.pkl")
print("title embeddings saved")
```

##BERT for F100 !!!

```python
import pandas as pd
from transformers import BertTokenizer, BertModel
import torch
import numpy as np

# טעינת המודל
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# טעינת הנתונים
df = pd.read_pickle("/content/drive/My Drive/Colab_Notebooks/data/full_embedding_15K.pkl")

# טיפול בנתונים בעמודת F100 - הפיכת כל הערכים למחרוזות והחלפת NaN
df['F100'] = df['F100'].astype(str).fillna("")

def get_bert_embedding(text, index, total):
    """ פונקציה שיוצרת Embeddings עם טיפול בשגיאות טקסט """
    if pd.isna(text) or not isinstance(text, str) or text.strip() == "":
        return np.zeros(768)  # מחזיר וקטור אפסים אם אין טקסט תקין

    if index % 500 == 0:
        print(f"Processing title {index}/{total}...")

    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=100)
    with torch.no_grad():
        output = model(**inputs)
    return output.last_hidden_state[:, 0, :].squeeze().numpy()

# גודל הצ'אנקים
chunk_size = 500

# עיבוד הנתונים בצ'אנקים
for i in range(0, len(df), chunk_size):
    chunk = df.iloc[i:i + chunk_size].copy()  # שימוש ב-copy() למניעת SettingWithCopyWarning

    # יצירת Embeddings
    chunk['bert_F100_embedding'] = [get_bert_embedding(text, j, len(chunk)) for j, text in enumerate(chunk['F100'])]

    # עדכון הנתונים חזרה ל-DataFrame הראשי
    df.loc[chunk.index, 'bert_F100_embedding'] = chunk['bert_F100_embedding']

    print(f"Processed chunk {i // chunk_size + 1} of {len(df) // chunk_size + 1}")

# שמירת הנתונים
df.to_pickle("/content/drive/My Drive/Colab_Notebooks/data/full_embedding_15K.pkl")
print("✅ Embeddings נשמרו בהצלחה!")
```
